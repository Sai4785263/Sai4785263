```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
#Fetch URL
def fetch_indeed_URL(what, where):
    URL = 'https://www.indeed.com/jobs?q={}&l={}&fromage=1'.format(what,where)
    return URL
def fetch_monster_URL(what, where):
    URL = 'https://www.linkedin.com/jobs/search/?f_TPR=r86400&keywords={}&location={}'.format(what,where)
    return URL
```

```python
def get_indeed_job_details(job_card):
    job_title = job_card.find('h2','jobTitle').text
    #print(job_title)
    job_company = job_card.find('span', 'companyName').text
    job_location = job_card.find('div', 'companyLocation').text
    job_link = 'https://www.indeed.com/' + job_card.get('href')
    job_descrp = job_card.find('div', 'job-snippet').text
    job_salary = job_card.find('span', 'estimated-salary')
    if job_salary != None:
        job_salary = job_salary.text
    job_postedDate = job_card.find('span', 'date').text
    return (job_title,job_company,job_location, job_link, job_descrp, job_salary, job_postedDate)

def get_linkedIn_job_details(job_card):
    print(job_card)
    job_title = job_card.find('h3','base-search-card__title').text
    job_company = job_card.find('a', 'hidden-nested-link').text
    job_location = job_card.find('span', 'job-search-card__location').text
    job_link = job_card.find("a",{"class":"hidden-nested-link"}).get("href")  
    print(job_link)
    #job_descrp = job_card.find('div', 'job-snippet').text
    job_salary = job_card.find('span', 'job-search-card__salary-info')
    if job_salary != None:
        job_salary = job_salary.text
    job_postedDate = job_card.find('time', 'job-search-card__listdate--new')
    if job_postedDate != None:
        job_postedDate = job_postedDate.text
    return (job_title,job_company,job_location, job_link, None, job_salary, job_postedDate)
```

```python
def fetch_indeed_jobs(what, where):
    indeed_url = fetch_indeed_URL(what, where)
    final_jobData = []
    for p in range(5):
        url=indeed_url+'&start={}'.format(p*10)
        print(url)
        response = requests.get(url)
        #print(indeed_url)
        soup = BeautifulSoup(response.text, 'html.parser')
        job_cards = soup.find_all('a', class_='tapItem')
        #print(job_cards)
        for each_card in job_cards:
            jobData = get_indeed_job_details(each_card)
            #print(jobData)
            final_jobData.append(jobData)
        p+=1
#     response = requests.get(indeed_url)
#     print(indeed_url)
#     soup = BeautifulSoup(response.text, 'html.parser')
#     job_cards = soup.find_all('a', class_='tapItem')
#     #print(job_cards)
#     final_jobData = []
#     for each_card in job_cards:
#         jobData = get_indeed_job_details(each_card)
#         #print(jobData)
#         final_jobData.append(jobData)
    print(final_jobData)
    indeed_jobs = pd.DataFrame(final_jobData)
    print(indeed_jobs.shape)
    return indeed_jobs

def fetch_linkedIn_jobs(what, where):
    monster_URL = fetch_monster_URL(what, where)
    response = requests.get(monster_URL)
    print(monster_URL)
    soup = BeautifulSoup(response.text, 'html.parser')
    job_cards = soup.find_all('div',class_='base-search-card__info')
    final_jobData = []
    for each_card in job_cards:
        jobData = get_linkedIn_job_details(each_card)
        #print(jobData)
        final_jobData.append(jobData)
        #print(final_jobData)
    monster_jobs = pd.DataFrame(final_jobData)
    return monster_jobs
```
